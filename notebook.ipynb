## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana perbandingan total pendapatan (monetary) di 5 negara bagian teratas antara periode Januari-September 2017 dan Januari-September 2018?

    - Specific: Fokus pada variabel total pendapatan (monetary) dan subjek geografis spesifik yaitu 5 negara bagian dengan performa tertinggi.
    - Measurable: Dapat diukur secara kuantitatif dengan menjumlahkan nilai transaksi (payment_value) dari data pembayaran.
    - Action-Oriented: Memberikan arahan bagi tim bisnis untuk memperkuat dominasi di wilayah sukses atau mengevaluasi strategi di wilayah yang pertumbuhannya melambat.
    - Relevant: Pendapatan adalah metrik kesehatan bisnis utama. Memahami sebaran geografis membantu efisiensi alokasi sumber daya.
    - Time-bound: Membandingkan dua rentang waktu yang setara: Januari-September 2017 dan Januari-September 2018.

- **Pertanyaan 2:** Kategori produk mana yang mengalami penurunan rata-rata skor ulasan paling signifikan pada periode Januari-September 2018 dibandingkan periode yang sama di tahun 2017?

    - Specific: Menargetkan identifikasi penurunan skor ulasan yang dikelompokkan berdasarkan kategori produk.
    - Measurable: Diukur dengan menghitung selisih rata-rata nilai dari 1.0 hingga 5.0 pada kolom review_score.
    - Action-Oriented: Mengarahkan tim operasional untuk melakukan investigasi kualitas pada kategori produk atau seller terkait guna mencegah churn pelanggan.
    - Relevant: Kepuasan pelanggan berdampak langsung pada reputasi platform dan retensi jangka panjang.
    - Time-bound: Dibatasi pada perbandingan performa Januari-September 2017 vs 2018.
    
- **Pertanyaan 3:** Apakah rata-rata waktu pengiriman (delivery time) mengalami efisiensi (menjadi lebih cepat) pada periode Januari-September 2018 dibandingkan periode yang sama di tahun 2017?

    - Specific: Berfokus pada durasi waktu pengiriman dari titik pembelian hingga barang diterima oleh pelanggan.
    - Measurable: Diukur dalam satuan hari dengan menghitung selisih antara order_purchase_timestamp dan order_delivered_customer_date.
    - Action-Oriented: Mengevaluasi kinerja mitra logistik; jika pengiriman melambat, platform perlu meninjau kembali kontrak kerja sama atau proses pergudangan.
    - Relevant: Kecepatan pengiriman adalah salah satu faktor penentu utama pengalaman belanja di e-commerce.
    - Time-bound: Fokus pada perbandingan periode Januari-September di kedua tahun tersebut.

- **Pertanyaan 4:** Bagaimana perbandingan distribusi frekuensi pembelian pelanggan (Frequency) antara periode Januari-September 2017 dan Januari-September 2018?

    - Specific: Menganalisis perilaku pembelian berulang (repeat order) oleh pelanggan unik.
    - Measurable: Diukur dengan menghitung jumlah order_id unik untuk setiap customer_unique_id dalam rentang waktu tersebut.
    - Action-Oriented: Jika distribusi frekuensi rendah (didominasi pembeli satu kali), maka dibutuhkan strategi re-marketing atau program loyalitas yang lebih agresif.
    - Relevant: Mengukur tingkat loyalitas pelanggan (aspek 'F' dalam RFM Analysis) sangat krusial untuk keberlanjutan bisnis.
    - Time-bound: Konsisten menggunakan perbandingan periode Januari-September 2017 dan 2018.

Note dalam membuat "Pertanyaan Bisnis"  
**Pendekatan Perbandingan Seimbang (Apple-to-Apple Comparison)**

Dalam proyek analisis ini, saya mengambil keputusan strategis untuk membatasi rentang waktu data yang dianalisis pada periode Januari hingga September untuk tahun 2017 dan 2018. Berikut adalah rincian pertimbangan dan alasan di balik pendekatan tersebut:

1. Keterbatasan Data (Data Constraints)
Berdasarkan hasil eksplorasi awal pada [orders_dataset.csv], ditemukan bahwa data transaksi tahun 2018 hanya tersedia hingga pertengahan bulan Oktober. Jika saya membandingkan data setahun penuh (12 bulan) pada 2017 dengan data yang tidak lengkap pada 2018, maka hasil agregasi (seperti total pendapatan atau volume penjualan) akan mengalami bias yang signifikan.

2. Menghindari Bias Musiman (Seasonality Bias)
Industri e-commerce sangat dipengaruhi oleh tren musiman, terutama pada kuartal keempat (Oktober–Desember) di mana terdapat lonjakan transaksi besar seperti Black Friday dan promo akhir tahun. Dengan memangkas data tahun 2017 hingga bulan September, saya memastikan bahwa lonjakan kuartal keempat di tahun 2017 tidak membuat performa tahun 2018 terlihat lebih rendah secara tidak adil.

3. Validitas Pertumbuhan (Year-to-Date Growth)
Dengan menggunakan rentang Bulan 1 - 9 pada kedua tahun tersebut, saya dapat melakukan analisis Year-to-Date (YTD). Hal ini memungkinkan saya untuk:

  - Menghitung tingkat pertumbuhan pendapatan secara akurat.
  - Membandingkan efisiensi logistik pada periode musim yang sama.
  - Melihat perubahan perilaku pelanggan secara objektif tanpa gangguan perbedaan durasi waktu.

4. Integritas Hasil Analisis
Pendekatan ini memastikan bahwa setiap kesimpulan dan rekomendasi aksi (action items) yang saya berikan kepada stakeholder didasarkan pada perbandingan yang valid dan tidak menyesatkan (misleading).

##### Melakukan Perbaikan Tabel (Order Reviews)

```python
# Mengisi nilai kosong pada judul dan pesan ulasan
order_reviews_df['review_comment_title'] = order_reviews_df['review_comment_title'].fillna("No Title")
order_reviews_df['review_comment_message'] = order_reviews_df['review_comment_message'].fillna("No Comment")

# Verifikasi hasil cleaning
print("Tipe data kolom tanggal:")
print(order_reviews_df[['review_creation_date', 'review_answer_timestamp']].dtypes)
print("\nJumlah missing values pada komentar:")
print(order_reviews_df[['review_comment_title', 'review_comment_message']].isna().sum())
```

**Insight:** (Opsional)
- Dengan mengonversi kolom review_creation_date dan review_answer_timestamp menjadi format datetime, kita sekarang dapat melakukan perhitungan berbasis waktu (time-series). 
- Mengisi nilai kosong (imputasi) pada review_comment_title dengan "No Title" dan review_comment_message dengan "No Comment" sangat membantu dalam menstandarisasi data. Hal ini memastikan bahwa proses analisis lanjutan (seperti filter teks atau analisis sentimen) tidak akan mengalami error akibat nilai NaN.

##### Melakukan Perbaikan Tabel (Products)

```python
# 1. Mengisi nilai kosong pada kolom kategori dengan "others"
products_df['product_category_name'] = products_df['product_category_name'].fillna("others")

# 2. Mengisi nilai kosong pada kolom atribut produk (nama, deskripsi, foto) dengan 0 atau "Unknown"
# Hal ini agar tipe data tetap konsisten (float/int)
products_df['product_name_lenght'] = products_df['product_name_lenght'].fillna(0)
products_df['product_description_lenght'] = products_df['product_description_lenght'].fillna(0)
products_df['product_photos_qty'] = products_df['product_photos_qty'].fillna(0)

# 3. Menangani nilai kosong pada kolom dimensi (hanya ada 2 baris)
# Karena jumlahnya sangat sedikit, kita bisa mengisinya dengan median agar tidak merusak distribusi
products_df['product_weight_g'] = products_df['product_weight_g'].fillna(products_df['product_weight_g'].median())
products_df['product_length_cm'] = products_df['product_length_cm'].fillna(products_df['product_length_cm'].median())
products_df['product_height_cm'] = products_df['product_height_cm'].fillna(products_df['product_height_cm'].median())
products_df['product_width_cm'] = products_df['product_width_cm'].fillna(products_df['product_width_cm'].median())

# Verifikasi hasil cleaning
print("Jumlah missing values setelah perbaikan:")
print(products_df.isna().sum())

products_df.info()
```

**Insight:** (Opsional)

```python
# IQR
def analyze_outliers(df, column, label):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[df[column] > upper_bound]
    percentage = (len(outliers) / len(df)) * 100
    
    print(f"=== ANALISIS OUTLIER: {label} ===")
    print(f"Batas Atas Wajar : {upper_bound:.2f}")
    print(f"Jumlah Outlier   : {len(outliers)} transaksi")
    print(f"Persentase       : {percentage:.2f}% dari total data")
    print(f"Nilai Maksimum   : {df[column].max():.2f}")
    print("-" * 30)
    return upper_bound

# 2. Analisis Outlier
upper_pv = analyze_outliers(main_revenue_df, 'payment_value', 'Nilai Pembayaran')
upper_dd = analyze_outliers(main_delivery_df, 'delivery_duration_days', 'Durasi Pengiriman')

# 3. Business Logic Check: Mencari item di mana biaya kirim lebih mahal dari harga produknya sendiri
expensive_freight = order_items_df[order_items_df['freight_value'] > order_items_df['price']]
print(f"=== BUSINESS LOGIC CHECK ===")
print(f"Item dengan Ongkir > Harga: {len(expensive_freight)} item ({ (len(expensive_freight)/len(order_items_df)*100):.2f}% )")
print("-" * 30)

# 4. Melihat Korelasi antara Durasi Pengiriman & Ulasan
plt.figure(figsize=(10, 6))
sns.boxplot(data=main_product_quality_df, x='review_score', y='delivery_duration_days', palette='viridis')
plt.ylim(0, 50) # Membatasi sumbu y agar fokus pada mayoritas data
plt.title('Hubungan Skor Ulasan dengan Durasi Pengiriman', fontsize=14)
plt.xlabel('Skor Ulasan (1-5)')
plt.ylabel('Durasi Pengiriman (Hari)')
plt.show()
```

- Outlier Pengiriman: Normalnya, barang sampai dalam maksimal 28 hari. Terdapat 3.708 pesanan yang mengalami keterlambatan parah (> 28 hari), dengan rekor terburuk mencapai 209 hari. Ini adalah isu operasional logistik yang harus segera diperbaiki!

```python
# Pertanyaan 3 : Apakah rata-rata waktu pengiriman (delivery time) mengalami efisiensi (menjadi lebih cepat) pada periode Januari-September 2018 dibandingkan periode yang sama di tahun 2017?

# 1. Filter periode Jan-Sep untuk 2017 dan 2018 dari dataframe yang sudah disiapkan
ytd_delivery_mask = (main_delivery_df['order_purchase_timestamp'].dt.month <= 9) & \
                    (main_delivery_df['order_purchase_timestamp'].dt.year.isin([2017, 2018]))

delivery_ytd = main_delivery_df[ytd_delivery_mask].copy()
delivery_ytd['year'] = delivery_ytd['order_purchase_timestamp'].dt.year

# 2. Hitung rata-rata waktu pengiriman per tahun
delivery_stats = delivery_ytd.groupby('year')['delivery_duration_days'].mean().reset_index()

# 3. Hitung selisih efisiensi
avg_2017 = delivery_stats.loc[delivery_stats['year'] == 2017, 'delivery_duration_days'].values[0]
avg_2018 = delivery_stats.loc[delivery_stats['year'] == 2018, 'delivery_duration_days'].values[0]
selisih = avg_2018 - avg_2017
status = "melambat" if selisih > 0 else "lebih cepat"

print(f"=== Analisis Efisiensi Pengiriman (Jan-Sep) ===")
print(f"Rata-rata pengiriman 2017: {avg_2017:.2f} hari")
print(f"Rata-rata pengiriman 2018: {avg_2018:.2f} hari")
print(f"Perubahan: {abs(selisih):.2f} hari ({status})")

# 4. Visualisasi Perbandingan
plt.figure(figsize=(8, 5))
colors = ['#4C72B0', '#C44E52'] if selisih > 0 else ['#4C72B0', '#55A868']

sns.barplot(data=delivery_stats, x='year', y='delivery_duration_days', palette=colors)

for i, v in enumerate(delivery_stats['delivery_duration_days']):
    plt.text(i, v + 0.2, f"{v:.2f} hari", ha='center', fontweight='bold')

plt.title('Perbandingan Rata-rata Waktu Pengiriman (2017 & 2018 Jan-Sep)', fontsize=13)
plt.ylabel('Durasi Pengiriman (Hari)')
plt.xlabel('Tahun')
plt.ylim(0, max(delivery_stats['delivery_duration_days']) + 3)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Pertanyaan 4 : Bagaimana perbandingan distribusi frekuensi pembelian pelanggan (Frequency) antara periode Januari-September 2017 dan Januari-September 2018?

# 1. Filter periode Jan-Sep untuk 2017 & 2018 dari main_revenue_df (yang sudah mengandung data customer)
ytd_freq_mask = (main_revenue_df['order_purchase_timestamp'].dt.month <= 9) & \
                (main_revenue_df['order_purchase_timestamp'].dt.year.isin([2017, 2018]))

df_ytd_freq = main_revenue_df[ytd_freq_mask].copy()
df_ytd_freq['year'] = df_ytd_freq['order_purchase_timestamp'].dt.year

# 2. Hitung jumlah pesanan per pelanggan unik per tahun
# Menggunakan customer_unique_id untuk mengidentifikasi pelanggan yang sama
freq_data = df_ytd_freq.groupby(['year', 'customer_unique_id']).agg({
    'order_id': 'count'
}).reset_index()
freq_data.rename(columns={'order_id': 'frequency'}, inplace=True)

# 3. Hitung distribusi jumlah pelanggan berdasarkan frekuensi pembeliannya
distribusi_freq = freq_data.groupby(['year', 'frequency']).size().reset_index(name='customer_count')

# 4. Visualisasi Distribusi 
plt.figure(figsize=(10, 6))
sns.barplot(
    data=distribusi_freq[distribusi_freq['frequency'] <= 3],
    x='frequency', 
    y='customer_count', 
    hue='year',
    palette='viridis'
)

plt.title('Distribusi Frekuensi Pembelian Pelanggan (YTD Jan-Sep)', fontsize=14)
plt.xlabel('Jumlah Pembelian (Frequency)', fontsize=12)
plt.ylabel('Jumlah Pelanggan Unik', fontsize=12)
plt.yscale('log') 
plt.legend(title='Tahun')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Output Ringkasan
print("=== Ringkasan Loyalitas Pelanggan (Jan-Sep) ===")
for yr in [2017, 2018]:
    total_cust = freq_data[freq_data['year'] == yr].shape[0]
    repeat_cust = freq_data[(freq_data['year'] == yr) & (freq_data['frequency'] > 1)].shape[0]
    percentage = (repeat_cust / total_cust) * 100
    print(f"Tahun {yr}: {repeat_cust} dari {total_cust} pelanggan melakukan pembelian ulang ({percentage:.2f}%)")
```

**Insight:** (Opsional)
- Meskipun jumlah pelanggan unik melonjak signifikan dari sekitar 26 ribu di 2017 menjadi lebih dari 52 ribu di 2018, pertumbuhan ini hampir seluruhnya didorong oleh akuisisi pelanggan baru. Dominasi mutlak pada kolom frekuensi "1" di grafik menunjukkan bahwa platform berhasil memperluas jangkauan pasarnya secara masif, namun masih sangat bergantung pada transaksi tunggal untuk mempertahankan volume penjualannya.
- Hal yang perlu diwaspadai adalah penurunan rasio pembelian ulang (repeat order) dari 6,19% menjadi 4,77%. Penurunan persentase ini di tengah ledakan jumlah pengguna mengindikasikan adanya fenomena "ember bocor", di mana upaya pemasaran sangat efektif mendatangkan orang baru tetapi kurang berhasil menjaga mereka untuk kembali. Secara operasional, ini menunjukkan bahwa kepuasan pelanggan atau program retensi belum mampu mengimbangi kecepatan ekspansi pasar.

## Analisis Lanjutan (Opsional)

```python
# PROSES RFM ANALYSIS (Recency, Frequency, Monetary)

# 1. Menyiapkan data dasar
# Kita butuh: customer_unique_id, order_purchase_timestamp, order_id, dan payment_value
rfm_df = main_revenue_df.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max", 
    "order_id": "nunique",            
    "payment_value": "sum"            
})

# 2. Mengganti nama kolom agar lebih representatif
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# 3. Menghitung Recency (Berapa hari sejak transaksi terakhir)
recent_date = main_revenue_df["order_purchase_timestamp"].max() + pd.Timedelta(days=1)
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

# Menghapus kolom timestamp karena sudah tidak diperlukan
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

# 4. Melakukan Scoring (Memberikan nilai 1-5 untuk setiap metrik)

# Menggunakan quantiles untuk membagi data menjadi 5 bagian sama besar
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

# Normalisasi rank ke skala 1-5
rfm_df['r_score'] = (rfm_df['r_rank'] / rfm_df['r_rank'].max() * 5).round(0)
rfm_df['f_score'] = (rfm_df['f_rank'] / rfm_df['f_rank'].max() * 5).round(0)
rfm_df['m_score'] = (rfm_df['m_rank'] / rfm_df['m_rank'].max() * 5).round(0)

# 5. Segmentasi Pelanggan

def segment_customer(df):
    if df['r_score'] >= 4 and df['f_score'] >= 4 and df['m_score'] >= 4:
        return 'Champions'
    elif df['r_score'] >= 3 and df['f_score'] >= 3:
        return 'Loyal Customers'
    elif df['r_score'] >= 3 and df['m_score'] >= 4:
        return 'Big Spenders'
    elif df['r_score'] <= 2:
        return 'At Risk / Hibernating'
    else:
        return 'Others'

rfm_df['customer_segment'] = rfm_df.apply(segment_customer, axis=1)

# ==============================================================================
# VISUALISASI HASIL RFM
# ==============================================================================

# Menghitung jumlah pelanggan di setiap segmen
segment_counts = rfm_df['customer_segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Customer Count']

plt.figure(figsize=(12, 6))
sns.barplot(
    data=segment_counts, 
    x='Customer Count', 
    y='Segment', 
    hue='Segment',
    palette='viridis',
    legend=False
)
plt.title('Distribusi Segmen Pelanggan Berdasarkan RFM', fontsize=15)
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Segmen')
plt.show()

# Menampilkan 5 data teratas hasil RFM
print("=== RINGKASAN DATA RFM ===")
print(rfm_df.head())
```

**Insight:**
- Metrik RFM 
    - Recency (R): Terlihat rentang yang sangat jauh, dari 161 hari hingga 586 hari. Pelanggan dengan nilai 586 hari (hampir 20 bulan) secara otomatis mendapatkan r_score rendah (0.0), yang memicu segmen "At Risk".
    - Frequency (F): Sesuai dengan temuan kita sebelumnya, nilai di kolom ini didominasi angka 1. Hal ini menyebabkan f_score cenderung seragam (2.0), sehingga pembeda utama antar segmen saat ini adalah seberapa baru mereka berbelanja dan seberapa banyak mereka membelanjakan uangnya.
    - Monetary (M): Nilai belanja bervariasi dari 27.19 hingga 196.89. Terlihat bahwa pelanggan dengan belanja rendah (27.19) mendapatkan m_score 0.0, yang berarti mereka berada di persentil bawah dalam hal kontribusi pendapatan.

- Analisis Segmen Pelanggan
    - At Risk / Hibernating (Row 2, 3, 4): Ini adalah kelompok "lampu kuning". Mereka pernah berbelanja dengan nilai yang lumayan (hingga 196.89), tetapi sudah tidak kembali selama lebih dari satu tahun (337 - 586 hari).
    - Others (Row 0, 1): Kelompok ini memiliki r_score yang lebih baik (4.0) karena kunjungan terakhir mereka lebih baru (~160 hari). Namun, karena frekuensi mereka masih 1, mereka belum bisa dikategorikan sebagai Loyal atau Champions.
